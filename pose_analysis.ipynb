### Analyze data

```python
import os
import pandas as pd
csv_path = os.path.join(videos_detected_dir, r'20210817-125617_points_090920211359.csv')
df = pd.read_csv(csv_path)
df
```

```python
x = df['backcenterx']
y = df['backcentery']
z = df['backcenterz']
topx = df['topcenterx']
topy = df['topcentery']
topz = df['topcenterz']
leftx = df['leftcenterx']
lefty = df['leftcentery']
leftz = df['leftcenterz']
rightx = df['rightcenterx']
righty = df['rightcentery']
rightz = df['rightcenterz']
time = list(range(len(x)))
import matplotlib.pyplot as plt
# plot
plt.figure(figsize=(10,10))
plt.plot(time,x, label='back-x')
plt.plot(time,y, label='back-y')
plt.plot(time,z, label='back-z')

plt.plot(time,topx, label='top-x')
plt.plot(time,topy, label='top-y')
plt.plot(time,topz, label='top-z')

plt.plot(time,leftx, label='left-x')
plt.plot(time,lefty, label='left-y')
plt.plot(time,leftz, label='left-z')

plt.plot(time,rightx, label='right-x')
plt.plot(time,righty, label='right-y')
plt.plot(time,rightz, label='right-z')
plt.gcf().autofmt_xdate()

plt.legend(loc='upper left')
plt.show()
```

#### Combine Cameras
You can choose to combine cameras with arithmetic mean (combination_type='mean') or averaging only the closest estimations determined by Euclidean distance(combination_type='similarity').

```python
import combining_cameras
combination_type='mean'
csv_name = 'length1'

df = combining_cameras.combine_cameras(csv_path, combination_type=combination_type)
output_csv_path = os.path.join(videos_detected_dir,f'{csv_name}_{combination_type}_combined.csv')
df.to_csv(path_or_buf=output_csv_path)
```

```python
x = df['finalCenterX']
y = df['finalCenterY']
z = df['finalCenterZ']
time = list(range(len(x)))
import matplotlib.pyplot as plt

# plot
plt.figure(figsize=(10,10))
plt.plot(time,x, label='x')
plt.plot(time,y, label='y')
plt.plot(time,z, label='z')

plt.gcf().autofmt_xdate()

plt.legend(loc='upper left')
plt.show()
```

#### Applying filters

```python
from scipy.signal import savgol_filter
window_length = 11 # The length of the filter window
polyorder = 2 # The order of the polynomial used to fit the samples. polyorder must be less than window_length.
#df.dropna(inplace=True) you can use this command if you wish to work only with detectable frames.

df = df.apply(lambda x: savgol_filter(x,window_length,polyorder) if ('X' in x.name or 'Y' in x.name or 'Z' in x.name) else x)
output_csv_path = os.path.join(videos_detected_dir,f'{csv_name}_{combination_type}_combined_savgol_filtered.csv')
df.to_csv(path_or_buf=output_csv_path)
```

```python
x = df['finalCenterX']
y = df['finalCenterY']
z = df['finalCenterZ']
time = list(range(len(x)))
import matplotlib.pyplot as plt

# plot
plt.figure(figsize=(10,10))
plt.plot(time,x, label='x')
plt.plot(time,y, label='y')
plt.plot(time,z, label='z')

plt.gcf().autofmt_xdate()

plt.legend(loc='upper left')
plt.show()
```
